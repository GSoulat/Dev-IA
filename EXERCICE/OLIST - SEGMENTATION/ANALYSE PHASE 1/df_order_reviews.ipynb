`Constat sur 99225 commentaire, il y a 87656 qui ne mettent pas de titre à leur commentaire et 58247 sans commentaire`

```python
SansTitre = df_order_reviews['review_comment_title'].isna().sum() / df_order_reviews.shape[0] * 100
print(f"il y a {SansTitre}% qui ne mettre pas de titre à leur commentaire")

SansCommentaire = df_order_reviews['review_comment_message'].isna().sum() / df_order_reviews.shape[0] * 100
print(f"il y a {SansCommentaire}% qui ne mettre pas de commentaire")


```
