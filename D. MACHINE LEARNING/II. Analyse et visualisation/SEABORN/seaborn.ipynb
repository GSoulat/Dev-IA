# FORMATION - UTILISATION DE SEABORN

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 

black_friday = pd.read_csv('BlackFriday.csv')
black_friday

```

# Boxplot

### pour ajouter un fond blank avec seaborn

Lien pour modifier le graphique [Lien seaborn](https://seaborn.pydata.org/generated/seaborn.axes_style.html#seaborn.axes_style)

```python
plt.rcParams["figure.figsize"]=[16,9] # Pour agrandir le graphique de façon global
with sns.axes_style('ticks', {'figure.facecolor': 'yellow', 'font.family': ['sans-serif'],'font.sans-serif': ['Arial',
  'DejaVu Sans'], 'axes.grid': True}):
    sns.boxplot(x="Gender", y="Purchase", data = black_friday, palette ='Set1')
    plt.draw()
```

```python
plt.rcParams["figure.figsize"]=[16,9] # Pour agrandir le graphique de façon global
with sns.axes_style('ticks', {'figure.facecolor': 'yellow', 'font.family': ['sans-serif'],'font.sans-serif': ['Arial',
  'DejaVu Sans'], 'axes.grid': True}):
    sns.boxplot(x="Gender", y="Purchase", hue = "Age" ,data = black_friday, palette ='muted')
    plt.legend(bbox_to_anchor= (1.05,1 ), loc= 2, borderaxespad= 0.) # Pour placer la légende à l'extérieur
    plt.draw()
```

# Countplot

```python
plt.rcParams["figure.figsize"]=[16,9] # Pour agrandir le graphique de façon global
with sns.axes_style('ticks', {'figure.facecolor': 'grey', 'font.family': ['sans-serif'],'font.sans-serif': ['Arial',
  'DejaVu Sans'], 'axes.grid': True}):
    sns.countplot(x="Age", hue="Gender", data = black_friday, palette = "Set1")
```

# Scatterplot catplot

```python
with sns.axes_style('ticks', {'figure.facecolor': 'grey', 'font.family': ['sans-serif'],'font.sans-serif': ['Arial',
  'DejaVu Sans'], 'axes.grid': True}):
    sns.catplot(x="Age", y="Purchase" , data = black_friday, palette = "Set1")
```

### la m&thode `sample()` prend aléatoirement un nombre de ligne fixé

```python
subset = black_friday.sample(1000)
```

```python
with sns.axes_style('ticks', {'figure.facecolor': 'grey', 'font.family': ['sans-serif'],'font.sans-serif': ['Arial',
  'DejaVu Sans'], 'axes.grid': True}):
    sns.catplot(x="Age", y="Purchase" , hue="Gender",data = subset, kind = "swarm")
    plt.draw()
```

# Faire des suplots avec Seannorn

```python
with sns.axes_style('ticks', {'figure.facecolor': 'grey', 'font.family': ['sans-serif'],'font.sans-serif': ['Arial',
  'DejaVu Sans'], 'axes.grid': True}):
  plt.subplot(211)
  sns.countplot(x="Age", hue="Gender", data = black_friday, palette = "Set1")
  plt.subplot(212)
  sns.boxplot(x="Gender", y="Purchase", hue = "Age" ,data = black_friday, palette ='muted')
  plt.legend(bbox_to_anchor= (1.05,1 ), loc= 2, borderaxespad= 0.) # Pour placer la légende à l'extérieur
  plt.draw()
```
