# FORMATION - UTILISATION DE MATPLOTLIB

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
black_friday = pd.read_csv('BlackFriday.csv')
black_friday
```

# CrÃ©ation d'un graphique vide

```python
plt.plot()
plt.draw()
```

```python
Age_groupe = black_friday.groupby('Age')
```

```python
Age_groupe.count()
```

```python
achats_clients = dict()
for age, group in Age_groupe:
    achats_clients[age] = sum(group['Purchase'])
    
achats_clients
```

```python
plt.plot(achats_clients.keys(), achats_clients.values())
plt.draw()
```

```python
plt.plot(achats_clients.keys(), achats_clients.values(), marker = "x",)
plt.draw()
```

```python
plt.plot(achats_clients.keys(), achats_clients.values(), marker = "x", linestyle = "")
plt.draw()
```

## Ajout de label

```python
plt.plot(achats_clients.keys(), achats_clients.values(), marker = "x", linestyle = "", color = "red")
plt.xlabel("CatÃ©gories d'age", color = "white", fontsize = 15, fontweight = "bold")
plt.ylabel("Somme totale des ventes", color = "white" , fontsize = 15, fontweight = "bold")
plt.title("Somme totale des ventes effectuÃ©es par catÃ©gories d'age", color = "white" , fontsize = 15, fontweight = "bold")
plt.draw()
```

# Pour changer la couleur par dÃ©faut avec la fonction `cmap`

## Pour changer le fonds du graphique de couleur

```python
plt.rcParams["figure.figsize"]=[16,9]

fig = plt.figure()
fig.patch.set_facecolor('grey') # Pour changer le fond couleur du graphique
plt.scatter(achats_clients.keys(), achats_clients.values(), c = list(achats_clients.values()), cmap="seismic")
plt.xlabel("CatÃ©gories d'age",fontsize = 15, fontweight = "bold")
plt.ylabel("Somme totale des ventes",  fontsize = 15, fontweight = "bold")
plt.title("Somme totale des ventes effectuÃ©es par catÃ©gories d'age", y=1.05 , fontsize = 15, fontweight = "bold") # place le titre Ã  une hauteur de 1.05 de plus par rapport au graphique
plt.draw()
```

## d'autre fonction cmap [link](https://matplotlib.org/stable/tutorials/colors/colormaps.html)

```python
fig = plt.figure()
fig.patch.set_facecolor('blue') # Pour changer le fond couleur du graphique
plt.scatter(achats_clients.keys(), achats_clients.values(), c = list(achats_clients.values()), cmap="seismic")
plt.xlabel("CatÃ©gories d'age",fontsize = 15, fontweight = "bold")
plt.ylabel("Somme totale des ventes",  fontsize = 15, fontweight = "bold")
plt.title("Somme totale des ventes effectuÃ©es par catÃ©gories d'age", y=1.05 , color="yellow", fontsize = 20, fontweight = "bold") # place le titre Ã  une hauteur de 1.05 de plus par rapport au graphique
plt.colorbar(orientation = "horizontal") # pour ajouter un bar de couleur en dessous du graphique
plt.draw()
```

```python
fig = plt.figure()
fig.patch.set_facecolor('blue') # Pour changer le fond couleur du graphique
plt.scatter(achats_clients.keys(), achats_clients.values(), c = list(achats_clients.values()), cmap="seismic", marker = "D")
plt.xlabel("CatÃ©gories d'age",fontsize = 15, fontweight = "bold")
plt.ylabel("Somme totale des ventes",  fontsize = 15, fontweight = "bold")
plt.title("Somme totale des ventes effectuÃ©es par catÃ©gories d'age", y=1.05 , color="yellow", fontsize = 20, fontweight = "bold") # place le titre Ã  une hauteur de 1.05 de plus par rapport au graphique
plt.colorbar(orientation = "vertical") # pour ajouter un bar de couleur en dessous du graphique
plt.draw()
```

## Camenbert (pie chart)

```python
data = black_friday["Age"].value_counts()
data
```

```python
plt.rcParams["figure.figsize"]=[9,9]
fig = plt.figure()
fig.patch.set_facecolor('grey') # Pour changer le fond couleur du graphique
plt.pie(data.values, labels=data.index, autopct='%1.1f%%', shadow=True, startangle=380)
plt.title("Somme totale des ventes effectuÃ©es par catÃ©gories d'age", y=1.05 , color="white", fontsize = 20, fontweight = "bold")
plt.draw()
```

## Mettre un graphique dans un graphique

```python
plt.rcParams["figure.figsize"]=[9,9]
fig = plt.figure()
fig.patch.set_facecolor('grey')
plt.subplot(211)
plt.bar(data.index, data,  color = ['blue','red'])
plt.xlabel("Sexe de l'acheteur",fontsize = 15, fontweight = "bold", color = 'blue')
plt.ylabel("Nombre d'acheteurs",  fontsize = 15, fontweight = "bold" , color = 'red')
plt.title("Somme totale des ventes effectuÃ©es par catÃ©gories d'age", y=1.05 , color="green", fontsize = 20, fontweight = "bold")

plt.subplot(212)
plt.boxplot(black_friday["Purchase"])
plt.xticks([1], ["Ventes"], color="red",fontsize = 20, fontweight = "bold")
plt.tight_layout()
plt.draw()
```
